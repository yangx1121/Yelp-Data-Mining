# Basic data cleaning

## Dealing the checkin time data

Since the data in the Check_in csv file is not in a readily useable format, let us make few data extraction according to a format that helps us to use this data to arrive at solutions.

```python
checkin['Monday'] = checkin.apply(lambda x: x[25:48].sum(), axis=1)
```

```python
checkin['Saturday'] = checkin.apply(lambda x: x[49:72].sum(), axis=1)
```

```python
checkin['Friday'] = checkin.apply(lambda x: x[1:24].sum(), axis=1)
```

```python
checkin['Sunday'] = checkin.apply(lambda x: x[73:96].sum(), axis=1)
```

```python
checkin['Thursday'] = checkin.apply(lambda x: x[97:120].sum(), axis=1)
```

```python
checkin['Tuesday'] = checkin.apply(lambda x: x[121:144].sum(), axis=1)
```

```python
checkin['Wednesday'] = checkin.apply(lambda x: x[145:168].sum(), axis=1)
```

```python
checkin['times'] = checkin.apply(lambda x: x[1:].sum(), axis=1)
```

```python
# Le us look how the "checkin" file looks like after the data cleaning process. 
checkin.head(10)
```

```python
# let us extract the columns from 'times' to 'Wednesday'.
df2=checkin.loc[:, 'times':'Wednesday']
df2.head(10)
```

```python
# Le us join both dataframes that we extractred in our earlier steps. 
checkin_times =pd.concat([df1,df2], axis=1)
```

```python
# Replicate value counts for each attribute that returns True/False answers (others must be done separately)

#all bakeries
s = pd.Series()
for each in att_list: 
    x = bake[each].value_counts()
    s = s.append(x)
    
#bakeries with reviews 4+ stars
s_ge4 = pd.Series()
for each in att_list: 
    x = bake_ge4[each].value_counts()
    s_ge4 = s_ge4.append(x)
    
#bakeries with reviews greater than 4 stars
s_gt4 = pd.Series()
for each in att_list: 
    x = bake_gt4[each].value_counts()
    s_gt4 = s_gt4.append(x)
    
#bakeries with reviews less than 4 stars
s_lt4 = pd.Series()
for each in att_list: 
    x = bake_lt4[each].value_counts()
    s_lt4 = s_lt4.append(x)
```

```python
#pull out True and False values for each...

#all bakeries
att_all_T = (s[1])
att_all_F = (s[0])

#bakeries with reviews 4+ stars
att_ge4_T = (s_ge4[1])
att_ge4_F = (s_ge4[0])
   
#bakeries with reviews greater than 4 stars
att_gt4_T = (s_gt4[1])
att_gt4_F = (s_gt4[0])

#bakeries with reviews less than 4 stars
att_lt4_T = (s_lt4[1])
att_lt4_F = (s_lt4[0])
```

```python
# make 'False' values into clean lists

#all bakeries
f = list()
for each in att_all_F:
    f.append(each)
    
#bakeries with reviews 4+ stars
f_ge4 = list()
for each in att_ge4_F:
    f_ge4.append(each)
    
#bakeries with reviews greater than 4 stars
f_gt4 = list()
for each in att_gt4_F:
    f_gt4.append(each)
    
#bakeries with reviews less than 4 stars
f_lt4 = list()
for each in att_lt4_F:
    f_lt4.append(each)
```

```python
# make 'True' values into clean lists

#all bakeries
t = list()
for each in att_all_T:
    t.append(each)
    
#bakeries with reviews 4+ stars
t_ge4 = list()
for each in att_ge4_T:
    t_ge4.append(each)
    
#bakeries with reviews greater than 4 stars
t_gt4 = list()
for each in att_gt4_T:
    t_gt4.append(each)
    
#bakeries with reviews less than 4 stars
t_lt4 = list()
for each in att_lt4_T:
    t_lt4.append(each)
```

```python
#check for missing values:

print(len(t))
print(len(f))

print(len(t_ge4))     #  1 missing value at position 7
print(len(f_ge4))

print(len(t_gt4))     #  3 missing values at positions 3,7,9
print(len(f_gt4))     #  1 missing value at position 27

print(len(t_lt4))     #  3 missing values at positions 2, 9, 11
print(len(f_lt4))     
```

```python
# insert zeros where counts were not returned:
missing_t_gt4 = [2,6,8]
for each in missing_t_gt4:
    t_gt4.insert(each, 0)

missing_f_gt4 = [26]
for each in missing_f_gt4:
    f_gt4.insert(each, 0)
    
missing_t_ge4 = [6]
for each in missing_t_ge4:
    t_ge4.insert(each, 0)
    
missing_t_lt4 = [1,8,10]
for each in missing_t_lt4:
    t_lt4.insert(each, 0)
```

```python
# Create database using Attribute names, True values, and False values
df_attall = pd.DataFrame(data = [att_list, t, f, t_lt4, f_lt4, t_ge4, f_ge4, t_gt4, f_gt4]).transpose()

#rename columns for clarity and add columns showing %s
df_attall.columns = ['Attribute', 'True_all', 'False_all', 'True_lt4', 'False_lt4', 'True_ge4', 'False_ge4', 'True_gt4', 'False_gt4']
df_attall['%all'] = df_attall['True_all']/(df_attall['True_all']+df_attall['False_all'])
df_attall['%lt4'] = df_attall['True_lt4']/(df_attall['True_lt4']+df_attall['False_lt4'])
df_attall['%ge4'] = df_attall['True_ge4']/(df_attall['True_ge4']+df_attall['False_ge4'])
df_attall['%gt4'] = df_attall['True_gt4']/(df_attall['True_gt4']+df_attall['False_gt4'])


df_attall
```

```python
# Run counts on non-binary attributes:

# Wifi analysis: surprisgingly not very strongly correlated with high reviews 
# (60% of all bakeries have free wifi, 62% of >4 star bakeries have free wifi)

all_wifi = bake['attributes.WiFi'].value_counts()
lt4_wifi = bake_lt4['attributes.WiFi'].value_counts()
ge4_wifi = bake_ge4['attributes.WiFi'].value_counts() 
gt4_wifi = bake_gt4['attributes.WiFi'].value_counts() 

levelnames = ["All", "lt4", "ge4", "gt4"]
Wifi = pd.concat([all_wifi, lt4_wifi, ge4_wifi, gt4_wifi], axis=1).transpose()
Wifi['Category'] = levelnames
Wifi['percent_freeWifi'] = Wifi['free']/(Wifi['free']+Wifi['no']+Wifi['paid'])

Wifi
```

```python
# Noise level analysis: Higher rated bakeries tend to be quieter
# (34% of all bakeries are quiet, but 40% of 4+star bakeries and 46% of 4.5-5star bakeries are quiet; 
# conversely, as ratings improve, percent of loud and even average noise level bakeries falls)

all_noise = bake['attributes.NoiseLevel'].value_counts() 
lt4_noise = bake_lt4['attributes.NoiseLevel'].value_counts() 
ge4_noise = bake_ge4['attributes.NoiseLevel'].value_counts() 
gt4_noise = bake_gt4['attributes.NoiseLevel'].value_counts() 

Noise = pd.concat([all_noise, lt4_noise, ge4_noise, gt4_noise], axis=1).transpose()
Noise['Category'] = levelnames
Noise['percent_average'] = Noise['average']/(Noise['average']+Noise['loud']+Noise['quiet']+Noise['very_loud'])
Noise['percent_loud'] = Noise['loud']/(Noise['average']+Noise['loud']+Noise['quiet']+Noise['very_loud'])
Noise['percent_quiet'] = Noise['quiet']/(Noise['average']+Noise['loud']+Noise['quiet']+Noise['very_loud'])
Noise['percent_very_loud'] = Noise['very_loud']/(Noise['average']+Noise['loud']+Noise['quiet']+Noise['very_loud'])

Noise
```
